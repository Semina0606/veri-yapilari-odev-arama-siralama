```python
import time

# Performans verilerini tutmak için global bir rapor sözlüğü
performans_raporu = {}

def performans_izleyici(algoritma_adi):
    """Algoritmaların hızını ölçen fonksiyonel sarmalayıcı."""
    def decorator(func):
        def wrapper(*args, **kwargs):
            baslangic = time.perf_counter()
            sonuc = func(*args, **kwargs)
            bitis = time.perf_counter()
            sure = (bitis - baslangic) * 1000
            performans_raporu[algoritma_adi] = f"{sure:.5f} ms"
            return sonuc
        return wrapper
    return decorator

# --- ARAMA ALGORİTMALARI ---

def dogrusal_arama(liste, hedef):
    """Linear Search: Veriyi baştan sona tarar."""
    for i in range(len(liste)):
        if liste[i] == hedef:
            return i
    return -1

def ikili_arama(sirali_liste, hedef):
    """Binary Search: Sıralı diziyi ikiye bölerek arar."""
    sol, sag = 0, len(sirali_liste) - 1
    while sol <= sag:
        orta = (sol + sag) // 2
        if sirali_liste[orta] == hedef:
            return orta
        elif sirali_liste[orta] < hedef:
            sol = orta + 1
        else:
            sag = orta - 1
    return -1

# --- SIRALAMA ALGORİTMALARI ---

@performans_izleyici("Kabarcık Sıralama (Bubble)")
def kabarcik_sirala(liste):
    n = len(liste)
    dizi = liste[:]
    for i in range(n):
        for j in range(0, n - i - 1):
            if dizi[j] > dizi[j + 1]:
                dizi[j], dizi[j + 1] = dizi[j + 1], dizi[j]
    return dizi

@performans_izleyici("Seçmeli Sıralama (Selection)")
def secmeli_sirala(liste):
    n = len(liste)
    dizi = liste[:]
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if dizi[j] < dizi[min_idx]:
                min_idx = j
        dizi[i], dizi[min_idx] = dizi[min_idx], dizi[i]
    return dizi

@performans_izleyici("Araya Ekleme (Insertion)")
def araya_ekleme_sirala(liste):
    dizi = liste[:]
    for i in range(1, len(dizi)):
        anahtar = dizi[i]
        j = i - 1
        while j >= 0 and anahtar < dizi[j]:
            dizi[j + 1] = dizi[j]
            j -= 1
        dizi[j + 1] = anahtar
    return dizi

@performans_izleyici("Hızlı Sıralama (Quick Sort)")
def hizli_sirala(liste):
    """Quick Sort: Pivot kullanarak hızlı sıralama yapar."""
    if len(liste) <= 1:
        return liste
    pivot = liste[len(liste) // 2]
    sol = [x for x in liste if x < pivot]
    orta = [x for x in liste if x == pivot]
    sag = [x for x in liste if x > pivot]
    return hizli_sirala(sol) + orta + hizli_sirala(sag)

# --- ÇALIŞTIRMA ---
if __name__ == "__main__":
    veri = [42, 12, 89, 23, 7, 56, 31, 4, 67, 10]
    print(f"Test Verisi: {veri}\n")

    # Sıralama Testleri
    s1 = kabarcik_sirala(veri)
    s2 = secmeli_sirala(veri)
    s3 = araya_ekleme_sirala(veri)
    s4 = hizli_sirala(veri) # Not: Quick sort recursive olduğu için decorator dışarıda tutulabilir veya sarmalanabilir.

    print("--- Algoritma Hız Raporu ---")
    for isim, sure in performans_raporu.items():
        print(f"{isim}: {sure}")

    # Arama Testleri
    hedef = 31
    print(f"\n'{hedef}' sayısı için arama sonuçları:")
    print(f"Doğrusal Arama: İndeks {dogrusal_arama(veri, hedef)}")
    print(f"İkili Arama (Sıralı): İndeks {ikili_arama(s4, hedef)}")
```
